```python
import pandas as pd 
url = '/content/Advertising.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)

data.plot(kind = 'scatter', x='TV', y = 'sales',figsize = (9,9),color='black')

from sklearn import linear_model
model = linear_model.LinearRegression()
X = pd.DataFrame(data['TV'])
model.fit(X,y = data['sales'])
print('Hệ số Intercept:',model.intercept_)
print('Hệ số Coeffient:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,y = data['sales']))

# data.plot(kind = 'scatter', x='TV', y = 'sales',figsize = (9,9),color='black')

y_pred = model.predict(X = pd.DataFrame(data['TV']))
plt.plot(data['TV'],y_pred,color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['TV'],data['TV']**2]).T 
model.fit(X,y = data['sales'])
print('Hệ số Intercept:',model.intercept_)
print('Hệ số Coeffient:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,y = data['sales']))

import numpy as np

range = np.arange(1.5,300,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize=(9,9),color='black',xlim=(0,300))
plt.plot(range, y_pred, color = 'blue')
```
