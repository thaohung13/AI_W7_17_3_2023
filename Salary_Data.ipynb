```python
import pandas as pd 
url = '/content/Salary_Data.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='YearsExperience', y = 'Salary',figsize = (9,9),color='black')

from sklearn import linear_model
model = linear_model.LinearRegression()
X = pd.DataFrame(data['YearsExperience'])
model.fit(X,y = data['Salary'])
print('Hệ số Intercept:',model.intercept_)
print('Hệ số Coeffient:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,y = data['Salary']))
```

```python
data.plot(kind = 'scatter', x='YearsExperience', y = 'Salary',figsize = (9,9),color='black')
y_pred = model.predict(X = pd.DataFrame(data['YearsExperience']))
plt.plot(data['YearsExperience'],y_pred,color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2]).T 
model.fit(X,y = data['Salary'])
print('Hệ số Intercept:',model.intercept_)
print('Hệ số Coeffient:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,y = data['Salary']))
```

```python
import numpy as np

range = np.arange(0,20,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'YearsExperience', y = 'YearsExperience', figsize=(9,9),color='blue')
plt.plot(range, y_pred, color = 'red')
```
