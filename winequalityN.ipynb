```python
import pandas as pd 
url = '/content/winequalityN.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='alcohol', y = 'quality',figsize = (9,9),color='black')

from sklearn import linear_model
model = linear_model.LinearRegression()
X = pd.DataFrame(data['alcohol'])
model.fit(X,y = data['quality'])
print('H s Intercept:',model.intercept_)
print('H s Coeffient:',model.coef_)
print(' chnh xc ca m hnh:',model.score(X,y = data['quality']))
```

```python
data.plot(kind = 'scatter', x='alcohol', y = 'quality',figsize = (9,9),color='black')
y_pred = model.predict(X = pd.DataFrame(data['alcohol']))
plt.plot(data['alcohol'],y_pred,color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['alcohol'],data['alcohol']**2]).T 
model.fit(X,y = data['quality'])
print('H s Intercept:',model.intercept_)
print('H s Coeffient:',model.coef_)
print(' chnh xc ca m hnh:',model.score(X,y = data['quality']))
```

```python
import numpy as np

range = np.arange(0,20,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'alcohol', y = 'alcohol', figsize=(9,9),color='blue')
plt.plot(range, y_pred, color = 'red')
```
