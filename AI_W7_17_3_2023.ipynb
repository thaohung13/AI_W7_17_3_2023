```python
import pandas as pd 
url = '/content/mtcars.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)

data.plot(kind = 'scatter', x='wt', y = 'mpg',figsize = (9,9),color='black')

from sklearn import linear_model
model = linear_model.LinearRegression()
X = pd.DataFrame(data['wt'])
model.fit(X,y = data['mpg'])
print('H s Intercept:',model.intercept_)
print('H s Coeffient:',model.coef_)
print(' chnh xc ca m hnh:',model.score(X,y = data['mpg']))

# data.plot(kind = 'scatter', x='wt', y = 'mpg',figsize = (9,9),color='black')

y_pred = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color = 'blue')
```

```python
model = linear_model.LinearRegression()
model.fit(X = data.loc[:,['wt','hp']],y=data['mpg'])

print('H s Intercept:', model.intercept_)
print('H s Coeffient:', model.coef_)
print(' chnh xc ca m hnh:',model.score(X = data.loc[:,['wt','hp']],y=data['mpg']))
```

```python
#Hi quy bc 1
from sklearn import linear_model
model=linear_model.LinearRegression()
#model.fit(): hm train 
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print('H s Intercept:', model.intercept_)
print('H s Coeffient:', model.coef_)
# chnh xc ca m hnh
print(' chnh xc ca m hnh:', model.score(X=pd.DataFrame(data['wt']),y=data['mpg']))
#y=-5,344x +  37,28
#Accuracy = 75%
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='r', xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
#Hi quy bc 2
import numpy as np

model = linear_model.LinearRegression()
X=pd.DataFrame([data['wt'],data['wt']**2]).T   #T: Transpose, chuyn v
model.fit(X,y=data['mpg'])
print("H s Intercept: ",model.intercept_)
print("H s coeffient l: ",model.coef_)
print(" chnh xc ca m hnh: ", model.score(X,y=data['mpg']))  
#y=-13,88x + 1,17.x^2 + 49,93
#Accuracy = 81,9%

range = np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color ='black', xlim=(0,7))
plt.plot(range, y_pred, color='blue')
```

```python
#hi quy bc 1
import pandas as pd 
url = '/content/50_Startups.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)

data.plot(kind = 'scatter', x='R&D Spend', y = 'Profit',figsize = (9,9),color='black')

from sklearn import linear_model
model = linear_model.LinearRegression()
X = pd.DataFrame(data['R&D Spend'])
model.fit(X ,y = data['Profit'])
print('H s Intercept:',model.intercept_)
print('H s Coeffient:',model.coef_)
print(' chnh xc ca m hnh:',model.score(X,y = data['Profit']))

# data.plot(kind = 'scatter', x='R&D Spend', y = 'Profit',figsize = (9,9),color='black')

y_pred = model.predict(X = pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'],y_pred,color = 'blue')
     
```
